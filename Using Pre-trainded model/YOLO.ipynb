```python
from moviepy.editor import VideoFileClip

# Input 3GP file path
input_file = "VID_20180211_102755.3gp"

# Output MP4 file path
output_file = "VID_20180211_102755.mp4"

# Load the video clip
clip = VideoFileClip(input_file)

# Define the new resolution and set it
new_resolution = (1280, 720)  # Change this to your desired resolution
clip_resized = clip.resize(new_resolution)

# Write the resized video to an MP4 file
clip_resized.write_videofile(output_file, codec="libx264")

print("Conversion complete.")
```
